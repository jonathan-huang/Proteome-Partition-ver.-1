# Plot for Special Function

```python
import numpy as np

def p_MM(c, k, m):
    term = 1 - c - m*k
    return (1 / (2*m)) * (term + np.sqrt(term**2 + 4*m*k))

def phi(alpha, beta, m):
    term = 1 - alpha - beta
    return (1 / (2*m)) * (term + np.sqrt(term**2 + 4*beta))
```

```python
import matplotlib.pyplot as plt

m = 0.5

# Create grid for alpha and beta on log scale
alpha = np.logspace(0, 5, 1000)  # 10^0 to 10^5
beta = np.logspace(0, 5, 1000)   # 10^0 to 10^5
Alpha, Beta = np.meshgrid(alpha, beta)

# Calculate phi values
Phi = phi(Alpha, Beta, m)

plt.figure(figsize=(10, 8))
contour = plt.contourf(Alpha, Beta, Phi, levels=50, cmap='viridis')
plt.colorbar(contour, label=r'$\phi(\alpha, \beta, m)$')
plt.contour(Alpha, Beta, Phi, levels=20, colors='black', alpha=0.3, linewidths=0.5)
plt.xlabel(r'$\alpha$', fontsize=14)
plt.ylabel(r'$\beta$', fontsize=14)
plt.title(r'Contour plot of $\phi(\alpha, \beta, m)$ with $m = 1/2$', fontsize=16)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, alpha=0.3, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

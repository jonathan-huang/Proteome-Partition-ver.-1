Analyzed under nutrient quality modulatio $b$.

## Optimization With Analytic Steady States

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.signal import savgol_filter
from joblib import Parallel, delayed
import pandas as pd
import time

# Constants
a1, a2 = 23.8, 1.4
# a1, a2 = 20, 10
# a1, a2 = 40, 2.8
# a1, a2 = 0.0001, 0.003
k1, k2 = 0.01, 0.003
# k1, k2 = 0.1, 0.1
N = 200

# Best range to test linear relationship
# b_min, b_max = 1e-3, 1e1

# b_min, b_max = 10, 300
# 25 crossover

# b_min, b_max, b_large = 1e-5, 5e-4, False   # small b
# b_min, b_max, b_large = 5e9, 1e10, True    # large b 
b_min, b_max = 1e-5, 300   # intermediate

b_vals = np.logspace(np.log10(b_min), np.log10(b_max), N)

b_threshold = 1e6
b_large = b_max > b_threshold
#b_vals = np.linspace(b_min, b_max, N+1)

# Approximate Theory
A = np.sqrt(k1/a1)+np.sqrt(k2/a2)
B = k1/a1 + k2/a2
C = (k1/a1)**(3/2) + (k2/a2)**(3/2)
D = (k1/a1)**2 + (k2/a2)**2
E = np.sqrt((k1*k2) / (a1*a2))
F = 1/a1 + 1/a2
```

## Softmax transformation

```python
# Softmax transformation from u1, u2 -> theta1, theta2, theta3 (LogSumExp)
from scipy.special import logsumexp
def softmax_partition(u):
    u1, u2 = u
    logZ = logsumexp([u1, 0.0, u2])
    theta1 = np.exp(u1 - logZ)
    theta2 = np.exp(-logZ)
    theta3 = np.exp(u2 - logZ)
    return theta1, theta2, theta3

# System
def BioSystemMM3(theta, b_val):
    term_Y1 = (a1 * theta[1]) / (b_val * theta[0]) + k1 - 1
    Y1 = 0.5 * (np.sqrt(term_Y1**2 + 4 * k1) - term_Y1)
    term_Y2 = (a2 * theta[2]) / (b_val * theta[0]) + k2 - (1 - Y1)
    Y2 = 0.5 * (np.sqrt(term_Y2**2 + 4 * (1 - Y1) * k2) - term_Y2)
    Y3 = (a2 * theta[2] / b_val) * (Y2 / (k2 + Y2))
    Y4 = (a2 * theta[1] * theta[2] / (b_val * theta[0])) * (Y2 / (k2 + Y2))
    Y5 = (a2 * theta[2]**2 / (b_val * theta[0])) * (Y2 / (k2 + Y2))
    lambda_val = b_val * Y3
    r_denom = ((k1 + Y1) / (a1 * Y4)) + ((k2 + Y2) / (a2 * Y5))
    r_avg = 1 / r_denom if r_denom != 0 else np.nan
    return np.array([b_val, theta[0], theta[1], theta[2], Y1, Y2, Y3, Y4, Y5, lambda_val, r_avg])

def objective_softmax(u, b_val): return -BioSystemMM3(softmax_partition(u), b_val)[9]   # objective function

# Optimization
def optimize_for_b_softmax(b_val, tol=1e-6, gtol=1e-6, maxiter=1000):
    if b_val > b_threshold:
        result = minimize(
            lambda u: objective_softmax(u, b_val),
            x0=[0.0, 0.0],
            bounds=[(-10, 10), (-10, 10)],
            method='L-BFGS-B',   # L-BFGS-B has bounds
            tol=tol,
            options={
                'gtol': gtol,
                'maxiter': maxiter,
                'disp': False
            }
        )
    else:
        result = minimize(
            lambda u: objective_softmax(u, b_val),
            x0=[0.0, 0.0],
            method='Powell',
            tol=tol,                       # global tolerance
            options={
                'xtol': tol,
                'ftol': tol,
                'maxiter': maxiter,        # maximum iterations
                'disp': False               # display output: True
            }
        )
    
    if result.success:
        return BioSystemMM3(softmax_partition(result.x), b_val)
    
    # Fallback to Nelder-Mead
    result_retry = minimize(
        lambda u: objective_softmax(u, b_val),
        x0=result.x,
        method='Nelder-Mead',
        tol=tol,
        options={'maxiter': maxiter, 'disp': False}
    )
    
    if result_retry.success:
        return BioSystemMM3(softmax_partition(result_retry.x), b_val)
        
    # Return best effort even if failed, to avoid gaps
    print(f"Warning: Optimization failed at b={b_val:.2e}")
    return BioSystemMM3(softmax_partition(result_retry.x), b_val)

# Replacing BioSystemMM3(theta1, theta2. theta3, b_val) with BioSystemMM3(array, b_val) increased
# performance by more than threefold! (2s -> 0.6 s) 20250924

# Parallel execution
start = time.time()
results = Parallel(n_jobs=-1)(delayed(optimize_for_b_softmax)(b) for b in b_vals)
results = np.array([r for r in results if r is not None])
print(f"Optimization completed in {time.time() - start:.2f} seconds.")

(b_arr, theta1_arr, theta2_arr, theta3_arr, 
 Y1_arr, Y2_arr, Y3_arr, Y4_arr, Y5_arr, 
 lambda_arr, r_avg_arr) = results.T   # The "Pythonic" way using tuple unpacking!
```

## Nanchaku Algorithm Implementation

```python
from dataclasses import dataclass

@dataclass
class Segment:
    start_idx: int
    end_idx: int
    slope: float
    intercept: float
    r_squared: float
    cv: float

def nunchaku_fixed_k(x, y, k_segments=3, min_points=5):
    """
    Forces the algorithm to partition the data into exactly 'k_segments'.
    Removes penalty logic to guarantee splitting.
    """
    # 1. Standardize (Keep this, it's good practice)
    y_mean, y_std = np.mean(y), np.std(y)
    if y_std == 0: y_std = 1.0
    y_scaled = (y - y_mean) / y_std
    
    n = len(x)
    costs = np.full((n, n), np.inf)
    
    # 2. Calculate Costs (SSE only, NO PENALTY)
    for i in range(n):
        # Ensure enough points
        for j in range(i + min_points, n):
            x_seg = x[i:j+1]
            y_seg_scaled = y_scaled[i:j+1]
            
            slope, intercept, _, _, _ = linregress(x_seg, y_seg_scaled)
            y_pred = slope * x_seg + intercept
            
            # Cost is just pure error
            costs[i, j] = np.sum((y_seg_scaled - y_pred)**2)

    # 3. Dynamic Programming (Force exactly k splits)
    # dp[k][i] = min cost to split first i points into k segments
    dp = np.full((k_segments + 1, n), np.inf)
    path = {} 

    # Initialize first segment
    for i in range(min_points, n):
        if costs[0, i] != np.inf:
            dp[1][i] = costs[0, i]
            path[(1, i)] = 0 

    # Iterate for 2..k segments
    for k in range(2, k_segments + 1):
        for i in range(min_points * k, n):
            # Try split point j
            for j in range(min_points * (k-1), i - min_points):
                if dp[k-1][j] != np.inf and costs[j+1, i] != np.inf:
                    total_cost = dp[k-1][j] + costs[j+1, i]
                    
                    if total_cost < dp[k][i]:
                        dp[k][i] = total_cost
                        path[(k, i)] = j + 1

    # 4. Reconstruct
    if dp[k_segments, n-1] == np.inf:
        print(f"Error: Could not force {k_segments} segments with min_points={min_points}.")
        return []

    segments = []
    curr_end = n - 1
    for k in range(k_segments, 0, -1):
        curr_start = path[(k, curr_end)]
        
        # Calculate real params for user
        x_seg = x[curr_start : curr_end+1]
        y_seg = y[curr_start : curr_end+1]
        slope, intercept, r_val, _, std_err = linregress(x_seg, y_seg)
        cv = (std_err / abs(slope)) if slope != 0 else np.inf
        
        segments.append(Segment(curr_start, curr_end, slope, intercept, r_val**2, cv))
        curr_end = curr_start - 1
    
    return list(reversed(segments))

# ================= Execution =================
# Ensure you use the sorted arrays from your data generation
# lam_sorted, Y5_sorted = ... (from your code)
sort_idx = np.argsort(lambda_arr)
lam_sorted = lambda_arr[sort_idx]
Y5_sorted = Y5_arr[sort_idx]

# Force exactly 3 segments
segments = nunchaku_fixed_k(lam_sorted, Y5_sorted, k_segments=3, min_points=10)

plt.figure(figsize=(10, 6))
plt.scatter(lam_sorted, Y5_sorted, s=10, color='gray', alpha=0.3)

for i, seg in enumerate(segments):
    x_seg = lam_sorted[seg.start_idx : seg.end_idx+1]
    y_fit = seg.slope * x_seg + seg.intercept
    
    label = f"Regime {i+1}"
    if i == 1: 
        label += " (Intermediate)"
        print(f"Intermediate Slope: {seg.slope}")
        print(f"Intermediate Range: {x_seg[0]:.4f} to {x_seg[-1]:.4f}")
        
    plt.plot(x_seg, y_fit, lw=3, label=label)

plt.legend()
plt.show()
```

```python
from scipy.optimize import brentq

A = np.sqrt(k1/a1) + np.sqrt(k2/a2)

# ------------------------------------------------------------------
# Large-b asymptotic constants from the three-sector analysis
# ------------------------------------------------------------------
# r_inf solves:
#   (k1 + 1 - r) (1 - r)^3 = (a2/(a1*k2)) * k1^2 * r^3
def r_inf_eq(r):
    return (k1 + 1.0 - r) * (1.0 - r)**3 - (a2/(a1 * k2)) * (k1**2) * r**3

r_inf = brentq(r_inf_eq, 1e-12, 1.0 - 1e-12)

eta_inf = np.sqrt(a2 * r_inf * (k1 + 1.0 - r_inf) / (a1 * k2 * (1.0 - r_inf)))
D_inf = k2 * (1.0 + eta_inf)**2 + r_inf

# Leading large-b optimizer:
# theta1 ~ s_inf / b,  theta2 ~ p_inf,  theta3 ~ q_inf
s_inf = a2 * (1.0 + eta_inf) / (eta_inf * D_inf)
p_inf = eta_inf * k2 * (1.0 + eta_inf) / D_inf
q_inf = (k2 * (1.0 + eta_inf) + r_inf) / D_inf

# ------------------------------------------------------------------
# Asymptotic approximation arrays
# ------------------------------------------------------------------
if b_large:
    theta1_approx_arr = s_inf / b_arr
    theta2_approx_arr = np.full_like(b_arr, p_inf, dtype=float)
    theta3_approx_arr = np.full_like(b_arr, q_inf, dtype=float)
else:
    # small-b approximation already in your code
    denom = 1 + 2*A*b_arr
    theta1_approx_arr = (1 + A*np.sqrt(b_arr)) / (1 + 2*A*np.sqrt(b_arr) + 2*A**2*b_arr)
    theta2_approx_arr = np.sqrt(k1/a1)*(1 - theta1_approx_arr) / A
    theta3_approx_arr = np.sqrt(k2/a2)*(1 - theta1_approx_arr) / A

# ------------------------------------------------------------------
# TPPA arrays from our three-sector formulas
# ------------------------------------------------------------------
sqrt_b = np.sqrt(b_arr)
denom_tppa = 1.0 + 0.5 * A**2 * b_arr + (A**2 / (2.0 * s_inf)) * b_arr**2

theta1_TPPA_arr = (1.0 - A * sqrt_b + 0.5 * A**2 * b_arr) / denom_tppa
theta2_TPPA_arr = (np.sqrt(k1/a1) * sqrt_b + p_inf * (A**2 / (2.0 * s_inf)) * b_arr**2) / denom_tppa
theta3_TPPA_arr = (np.sqrt(k2/a2) * sqrt_b + q_inf * (A**2 / (2.0 * s_inf)) * b_arr**2) / denom_tppa

# ------------------------------------------------------------------
# Plots
# ------------------------------------------------------------------

# Partition component theta3 to nutrient levels b
plt.figure(figsize=(8,6))
plt.plot(b_arr, theta1_arr, lw=2, label=r'$\theta_1$', marker='o', color='red')
plt.plot(b_arr, theta1_approx_arr, lw=2, label=r'$\theta_1$ theory', color='blue')
plt.xlabel(r'$b$', fontsize=label_size); plt.ylabel(r'$\theta_1$', fontsize=label_size)
plt.title(r'Partition component ($\theta_1$) vs nutrient levels $b$', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.grid(True)
# plt.ylim(0.0, 0.01)
plt.legend()
plt.show()

# Partition component theta2 to nutrient levels b
plt.figure(figsize=(8,6))
plt.plot(b_arr, theta2_arr, lw=2, label=r'$\theta_2$', marker='o', color='orange')
plt.plot(b_arr, theta2_approx_arr, lw=2, label=r'$\theta_2$ theory', color='blue')
plt.xlabel(r'$b$', fontsize=label_size); plt.ylabel(r'$\theta_2$', fontsize=label_size)
plt.title(r'Partition component ($\theta_2$) vs nutrient levels $b$', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.grid(True)
# plt.ylim(0.0, 0.01)
plt.legend()
plt.show()

# Partition component theta3 to nutrient levels b
plt.figure(figsize=(8,6))
plt.plot(b_arr, theta3_arr, lw=2, label=r'$\theta_3$', marker='o', color='yellow')
plt.plot(b_arr, theta3_approx_arr, lw=2, label=r'$\theta_3$ theory', color='blue')
plt.xlabel(r'$b$', fontsize=label_size); plt.ylabel(r'$\theta_3$', fontsize=label_size)
plt.title(r'Partition component ($\theta_3$) vs nutrient levels $b$', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.grid(True)
# plt.ylim(0.0, 0.01)
plt.legend()
plt.show()

# Growth rate to protein partition \theta
plt.figure(figsize=(8,6))
plt.scatter(theta1_arr, lambda_arr, label=r'$\theta_1$', c='red', alpha=0.7)
plt.scatter(theta2_arr, lambda_arr, label=r'$\theta_2$', c='orange', alpha=0.7)
plt.scatter(theta3_arr, lambda_arr, label=r'$\theta_3$', c='yellow', alpha=0.7)
plt.xlabel(r'Protein partition ($\theta$)', fontsize=label_size)
plt.ylabel(r'Growth rate ($\lambda$)', fontsize=label_size)
plt.title(r'Growth rate ($\lambda$) vs protein partition ($\theta$)', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# Normalized plot growth rate to protein partition \theta
plt.figure(figsize=(8,6))
plt.scatter(theta1_arr, np.log(b_arr), label=r'$\theta_1$', c='red', alpha=0.7)
plt.scatter(theta2_arr, np.log(b_arr), label=r'$\theta_2$', c='orange', alpha=0.7)
plt.scatter(theta3_arr, np.log(b_arr), label=r'$\theta_3$', c='yellow', alpha=0.7)
plt.scatter(lambda_arr/(lambda_arr.max()),np.log(b_arr), label=r'Normalized growth rate', c='darkred', marker='x')
plt.xlabel(r'Component value ($\theta$ or normalized $\lambda$)', fontsize=label_size)
plt.ylabel(r'Nutrient level ($b$)', fontsize=label_size)
plt.title(r'Nutrient levels ($b$) vs protein partitions ($\theta$) and growth rate ($\lambda$)', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# ------------------------------------------------------------------
# Diagnostic tests for the large-b regime
# ------------------------------------------------------------------

if b_large:
    # Large-b asymptotic growth-rate limit from the theory
    lambda_inf = a2 * r_inf / D_inf

    # Sort by b for cleaner lines
    idx = np.argsort(b_arr)
    b_sorted = b_arr[idx]

    theta1_sorted = theta1_arr[idx]
    theta2_sorted = theta2_arr[idx]
    theta3_sorted = theta3_arr[idx]
    lambda_sorted = lambda_arr[idx]

    theta1_approx_sorted = theta1_approx_arr[idx]
    theta2_approx_sorted = theta2_approx_arr[idx]
    theta3_approx_sorted = theta3_approx_arr[idx]

    theta1_TPPA_sorted = theta1_TPPA_arr[idx]
    theta2_TPPA_sorted = theta2_TPPA_arr[idx]
    theta3_TPPA_sorted = theta3_TPPA_arr[idx]

    # ==========================================================
    # Test 1: plot b * theta1 vs b
    # If theta1 ~ s_inf / b, then b*theta1 -> s_inf
    # If theta1 -> 1/3, then b*theta1 grows ~ b/3
    # ==========================================================
    plt.figure(figsize=(8,6))
    plt.plot(b_sorted, b_sorted * theta1_sorted, 'o', ms=5, alpha=0.7,
            label=r'Numerical $b\theta_1$', color='red')
    plt.plot(b_sorted, b_sorted * theta1_approx_sorted, lw=2,
            label=r'Large-$b$ asymptotic $b\theta_1$', color='blue')
    plt.plot(b_sorted, b_sorted * theta1_TPPA_sorted, lw=2, ls='--',
            label=r'TPPA $b\theta_1$', color='purple')
    plt.axhline(s_inf, color='black', ls=':', lw=2,
                label=rf'Theory limit $s_\infty={s_inf:.4g}$')
    plt.xscale('log')
    plt.xlabel(r'$b$', fontsize=label_size)
    plt.ylabel(r'$b\,\theta_1$', fontsize=label_size)
    plt.title(r'Test 1: does $b\theta_1$ approach a constant?', fontsize=title_size)
    plt.tick_params(axis='both', which='major', labelsize=tick_size)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Optional: compare b*theta1 against the equipartition prediction b/3
    plt.figure(figsize=(8,6))
    plt.plot(b_sorted, b_sorted * theta1_sorted, 'o', ms=5, alpha=0.7,
            label=r'Numerical $b\theta_1$', color='red')
    plt.plot(b_sorted, b_sorted / 3.0, lw=2, ls='--',
            label=r'Equipartition prediction $b/3$', color='green')
    plt.axhline(s_inf, color='black', ls=':', lw=2,
                label=rf'Boundary-layer prediction $s_\infty={s_inf:.4g}$')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$b$', fontsize=label_size)
    plt.ylabel(r'$b\,\theta_1$', fontsize=label_size)
    plt.title(r'Comparison: equipartition vs boundary-layer scaling', fontsize=title_size)
    plt.tick_params(axis='both', which='major', labelsize=tick_size)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
from scipy.optimize import minimize_scalar

b_contour = 5.0

N_grid = 100
t3_vals = np.linspace(0.001, 0.999, N_grid)
t1_vals = np.linspace(0.001, 0.999, N_grid)
T3, T1 = np.meshgrid(t3_vals, t1_vals)
Lambda_grid = np.full_like(T3, np.nan)

for i in range(N_grid):
    for j in range(N_grid):
        t3 = T3[i, j]
        t1 = T1[i, j]
        t2 = 1.0 - t1 - t3
        if t2 > 0:
            res = BioSystemMM3([t1, t2, t3], b_contour)
            Lambda_grid[i, j] = res[9]

line_t3 = np.linspace(0.01, 0.99, 50)
line_t1 = []

for t3_fixed in line_t3:
    # Maximize lambda(t1) subject to 0 <= t1 <= 1 - t3_fixed
    # theta2 = 1 - t1 - t3_fixed
    
    def objective(t1):
        t2 = 1.0 - t1 - t3_fixed
        if t2 <= 0: return 1e10 # Penalty
        res = BioSystemMM3([t1, t2, t3_fixed], b_contour)
        return -res[9] # Negative for minimization

    upper_bound = 1.0 - t3_fixed - 1e-7 # Small epsilon to avoid t2=0
    if upper_bound < 0: upper_bound = 0
    
    res = minimize_scalar(objective, bounds=(0, upper_bound), method='bounded')
    
    if res.success:
        line_t1.append(res.x)
    else:
        line_t1.append(np.nan)

# 4. Plot
plt.figure(figsize=(8, 6))
ax = plt.gca()
ax.set_facecolor('lightgrey') # Color the invalid region (where theta1 + theta3 > 1) grey

# Contour
cp = plt.contourf(T3, T1, Lambda_grid, levels=50, cmap='viridis')
cbar = plt.colorbar(cp)
cbar.set_label(r'Growth rate $\lambda$', fontsize=label_size)
cbar.ax.tick_params(labelsize=tick_size)

# Optimal Line
plt.plot(line_t3, line_t1, 'w--', linewidth=2.5, label=r'Optimal $\theta_1$ for fixed $\theta_3$' + '\n' + r'at $b={:.1f}$'.format(b_contour))

# Labels
plt.xlabel(r'$\theta_3$ (Ribosomal protein partition)', fontsize=label_size)
plt.ylabel(r'$\theta_1$ (Transport protein partition)', fontsize=label_size)
plt.title(f'Growth Rate Landscape', fontsize=title_size)
plt.tick_params(axis='both', which='major', labelsize=tick_size)
plt.legend(loc='upper right', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
